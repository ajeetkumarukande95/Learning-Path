```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
tokenizer=AutoTokenizer.from_pretrained('gpt2')
model=AutoModelForCausalLM.from_pretrained('gpt2')
```

```python
# greedy search
input = tokenizer("this is a nlp class",return_tensors="pt")
output=model.generate(**input)
```

```python
input
```

```python
output
```

```python
tokenizer.decode(output[0])
```

```python
# beam search
input=tokenizer('this is nlp class',return_tensors='pt')
output=model.generate(**input,num_beams=6)
tokenizer.decode(output[0])
```

```python
## no_repeat_n_gram
```

```python
input=tokenizer("this is a nlp class",return_tensors='pt')
output=model.generate(**input,num_beams=6,max_new_tokens=50)
tokenizer.decode(output[0])
```

```python
input = tokenizer("this is nlp class",return_tensors='pt')
output = model.generate(**input,num_beams=6,max_new_tokens=50)
tokenizer.decode(output[0])
```

```python
input = tokenizer("this is a nlp class", return_tensors="pt")

output = model.generate(**input, num_beams=6, max_new_tokens=50, no_repeat_ngram_size=2)

tokenizer.decode(output[0])
```

```python
# random sampling
input=tokenizer("this is a nlp class",return_tensors='pt')
output=model.generate(**input,num_beams=6,max_new_tokens=50,no_repeat_ngram_size=2,do_sample=True,temperature=0.9)
tokenizer.decode(output[0])
```

```python
# temperature
```

```python
## temperature
input = tokenizer("this is a nlp class", return_tensors="pt")

output = model.generate(**input, num_beams=6, max_new_tokens=50, no_repeat_ngram_size=2, do_sample=True, temperature=0.1)

tokenizer.decode(output[0])
```

```python
## top_k --> next word chosen from top k token
```

```python
## top_k
input = tokenizer("this is a nlp class", return_tensors="pt")

output = model.generate(**input, num_beams=6, max_new_tokens=50, no_repeat_ngram_size=2, do_sample=True, top_k=5)

tokenizer.decode(output[0])
```

```python
## top_p --> next word chosen from tokens whose sum probability equal to top_p
```

```python
## top_p
input = tokenizer("this is a nlp class", return_tensors="pt")

output = model.generate(**input, num_beams=6, max_new_tokens=50, no_repeat_ngram_size=2, do_sample=True, top_p=0.8)

tokenizer.decode(output[0])
```
