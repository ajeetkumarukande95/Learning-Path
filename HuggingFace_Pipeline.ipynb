```python
from transformers import pipeline
```

```python
# summerizer = pipeline('summarization', model='google-t5/t5-base')
summerizer = pipeline('summarization',model='google-t5/t5-small')
```

```python
text = '''Hugging Face Hub supports all file formats, but has built-in features for GGUF format, a binary format that is optimized for
quick loading and saving of models, making it highly efficient for inference purposes. GGUF is designed for use with GGML and other
executors. GGUF was developed by @ggerganov who is also the developer of llama.cpp, a popular C/C++ LLM inference framework.
Models initially developed in frameworks like PyTorch can be converted to GGUF format for use with those engines.'''
```

```python
summerizer(text)
```

```python
sumaerized_text = summerizer(text)
```

```python
sumaerized_text[0]['summary_text']
```

```python
#sentiment_classifer = pipeline("sentiment-analysis")
sentiment_classifier = pipeline("sentiment-analysis")
```

```python
sentiment_classifier(sumaerized_text[0]['summary_text'])
```

```python
sentiment_classifier("this product is great.")
```

```python
text_list = ["this product is great.",
             "it sucks",
             "it is expensive",
             "good quality",
             "product is worst",
             "this is a product of company"]

sentiment_classifier(text_list)
```

```python
#zero_classifier
zero_classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
zero_classifier("My job is to perform preprocessing on text and build models",
                candidate_labels=["Machine learning", "deep learning", "NLP"])
```
